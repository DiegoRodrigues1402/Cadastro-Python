```python
import requests
import json
import pandas as pd
```

```python
dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
```

```python
resultado = json.loads(dados_frutas.text)
```

```python
dados = pd.DataFrame(resultado)
dados.head(10)
```

```python
pd.json_normalize(dados['nutritions'])
```

```python
dados.to_json('arquivo_frutas.json')
```
