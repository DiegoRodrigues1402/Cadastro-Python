```python
import pandas as pd
```

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'#usamos esse comando para carregar arquivo xlsx
```

```python
dados = pd.read_excel(url)
```

```python
dados.head(5)
```

```python
pd.ExcelFile(url).sheet_names#usamos esse comando para retornar o nome das paginas se existir
```

```python
percapita = pd.read_excel(url,sheet_name = 'emissoes_percapita')#com esse comando sheet_name = utilizamos para visualizar pagina selecionada
percapita
```

```python
fontes = pd.read_excel(url,sheet_name = 'fontes')#selecionar ultima fonte
fontes
```

```python
co2 = pd.read_excel(url,sheet_name = 'emissoes_C02',usecols='a:d')#usamos esse comando usecols= para visualizar colunas que gostariamos visualizar podemos usar dessa forma ou
#selecionar pagina a pagina
co2
```

```python
co2 = pd.read_excel(url,sheet_name = 'emissoes_C02',usecols=[0,1,2,3])#podemos usar pelo indice da pagina
co2
```

```python
co2_2 = pd.read_excel(url,sheet_name = 'emissoes_C02',usecols='a:d',nrows=10)#com esse comando nrows=10 selecionamos quantidade de linhas a ser exibida
co2_2
```
